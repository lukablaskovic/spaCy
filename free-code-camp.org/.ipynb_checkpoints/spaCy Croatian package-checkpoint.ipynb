```python
import spacy
nlp = spacy.load("hr_core_news_sm")
from spacy import displacy
```

```python
with open ("text.txt", "r") as f:
    text= f.read()
```

```python
doc = nlp(text)
```

```python
#Convert generator into list
sen1 = list(doc.sents)[0]
sen2 = list(doc.sents)[1]
print(sen1)
print(sen2)
```

```python
displacy.render(doc, style="dep")
```

```python
#Small model (doesnt contain word vectors), makes some mistakes
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
displacy.render(doc, style="ent")
```

```python
for token in doc[:10]:
    print(token)
```

```python
import spacy
from spacy import displacy
nlp = spacy.load("hr_core_news_lg")
text = """Pozdrav, zanima me koliko će me koštati upis više godine s obzirom da nisam položio dva kolegija od 12 ECTS bodova. Dodatno, do kada moram uplatiti navedeni iznos?"""
doc = nlp(text)
displacy.render(doc, style="dep")
```
